```python
import numpy as np 
from scipy.stats import gamma
```

```python
shape, scale = 2,2
s= np.random.gamma(shape, scale, 1000)
```

```python
print(s)
```

```python
shape, scale, loc = gamma.fit(s, floc=0)
```

```python
print(f" Estimated shape : {shape}")
print(f" Estimated scale: {scale}")
```
